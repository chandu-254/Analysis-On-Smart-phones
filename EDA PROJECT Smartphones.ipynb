## Handling the Missing Values

```python
df=df.copy()
```

```python
df['Memory'].dtypes
```

```python

df['Memory']=df['Memory'].fillna(df['Memory'].mode()[0])
```

```python
df['Storage'].dtypes
```

```python
df['Storage']=df['Storage'].fillna(df['Storage'].mode()[0])
```

```python
df['Rating'].dtypes
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.What is the distribution of smartphone prices?
```

```python
fig=plt.figure(figsize=(4,4),facecolor='skyblue')
ax=plt.axes(facecolor='lavender')
sns.histplot(x=df['Selling Price'],bins=10,color='purple',label='Price',kde=True)
plt.title('Distribution of Smartphone Prices')
plt.xlabel('Selling Price')
plt.ylabel('customer count')
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(13,6),facecolor='deepskyblue')
ax=plt.axes(facecolor='lightcyan')
plt.bar(df['Brands'].value_counts().index,df['Brands'].value_counts(),label='Brands',color='grey')
plt.xlabel('Brands')
plt.ylabel('total of each ')
plt.title('Number of Smart Phones by Brand')
plt.legend()
plt.show()
```

```python
a=df.groupby('Brands')['Mobile'].agg('count')
a
```

```python

sns.boxplot(data=df,y='Memory')
plt.title('Distribution of RAM Capacity')
plt.show()
```

```python
df.groupby('Memory')['Mobile'].count()
```

## Bivariate Analysis categorical and categorical

```python
fig=plt.figure(figsize=(16,6))
sns.scatterplot(df,y=df['Storage'],x=df['Memory'])
plt.xlabel('RAM')
plt.ylabel('Storage')
plt.title('Relationship between RAM and Storage' )
plt.show()
```

## Bivariate Analysis Categorical vs Numerical

```python
plt.figure(figsize=(14,6))
plt.axes(facecolor='honeydew')
sns.boxplot(data=df,x='Brands',y='Selling Price',color='aqua')
plt.title('Price Comparision of each Brand')
plt.show()
```

```python
df.groupby('Brands')['Selling Price'].mean()
```

```python
df.groupby('Brands')['Selling Price'].median()
```

## Bivariate Analysis Numeric and Numeric

```python

sns.scatterplot(data=df,x='Selling Price',y='Rating',hue='Brands')
plt.title('Relationship between Price and rating')
```

```python
## Multivariate Analysis
```

```python
df[['Original Price',"Selling Price","Rating"]].corr()
```

```python
sns.heatmap(df[['Original Price',"Selling Price","Rating"]].corr(),annot=True)
plt.title('Correlation of Price and rating')
plt.show()
```

```python
sns.pairplot(df[['Rating','Original Price','Selling Price']])
plt.suptitle('Relationship Between SmartPhone Features')
plt.show()
```

```python
df[['Selling Price','Discount','discount percentage']].corr()
```
